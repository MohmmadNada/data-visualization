Import part

```python
import seaborn as sns # import part 
import matplotlib.pyplot as plt
```

```python
sns.set_style("ticks")
carCrashesData=sns.load_dataset('car_crashes').head(10)
# carCrashesData.head(5)
```

```python
# waht is the most effect factor in car crashes ? alcohol or speed?
sns.regplot(x="total", y="speeding", data=carCrashesData);
sns.regplot(x="total", y="alcohol", data=carCrashesData);# blue
```

### Conclusion :   alcohol has more slope and total death , so its more effect
